```python
# Vòng lặp này lặp qua danh sách DataFrames (VCB, SSI, FPT, DP3, FRT, VCG). Các DataFrame này chứa thông tin về các cổ phiếu khác nhau
for stock_df in [VCB, SSI, FPT, DP3, FRT, VCG]:
    if stock_df['Price'].dtype == 'O':
        stock_df.loc[:, 'Price'] = stock_df['Price'].str.replace(',', '', regex=True) 
        stock_df.loc[:, 'Price'] = pd.to_numeric(stock_df['Price'], errors='coerce')
        # Phần này kiểm tra xem cột 'Price' trong mỗi DataFrame có kiểu dữ liệu 'O' (object) hay không, điều này thường chỉ ra rằng nó có thể chứa các ký tự không phải số. Nếu vậy, nó sẽ xóa dấu phẩy khỏi cột 'Price' bằng cách sử dụng str.replace và chuyển đổi các giá trị thành số bằng cách sử dụng pd.to_numeric. errors='coerce' được sử dụng để thay thế mọi giá trị không thể chuyển đổi bằng NaN

# Ghép nối các khung dữ liệu cổ phiếu
stocks = pd.concat([VCB['Price'], SSI['Price'], FPT['Price'], DP3['Price'], FRT['Price'], VCG['Price']], axis=1)

# Đổi tên cột lần lượt đúng với các dữ liệu cổ phiếu
stocks.columns = ['VCB', 'SSI', 'FPT', 'DP3', 'FRT', 'VCG']
```

MA TRẬN HIỆP PHƯƠNG SAI

```python
# Log của phần trăm thay đổi
cov_matrix = stocks.pct_change().apply(lambda x: np.log(1+x)).cov()
cov_matrix
```

LỢI NHUẬN KỲ VỌNG DANH MỤC ĐẦU TƯ

```python
# Sử dụng hàm resample() để nhận lợi nhuận hàng năm. Đối số của hàm, 'Y', biểu thị hàng năm.
ind_er = stocks.resample('Y').last().pct_change().mean()
ind_er
```

TỐI ƯU HÓA DANH MỤC ĐẦU TƯ BẰNG CÁCH SỬ DỤNG ĐƯỜNG BIÊN HIỆU QUẢ

```python
# Độ biến động được tính bằng độ lệch chuẩn hàng năm. Tôi nhân với 250 vì thường có 250 ngày giao dịch/năm.
ann_sd = stocks.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(250))
ann_sd
```

```python
assets = pd.concat([ind_er, ann_sd], axis=1) # Tạo bảng để trực quan hóa lợi nhuận và sự biến động của tài sản
assets.columns = ['Returns', 'Volatility']
assets
```

```python
p_ret = [] # Xác định một mảng trống cho lợi nhuận danh mục đầu tư
p_vol = [] # Xác định một mảng trống cho sự biến động của danh mục đầu tư
p_weights = [] # Xác định một mảng trống cho trọng số tài sản (asset weights)

# Chạy vòng lặp này 10000 lần
num_assets = len(stocks.columns)
num_portfolios = 10000
```

```python
np.random.seed(42)
# Đảm bảo khả năng tái tạo, nghĩa là nếu tôi chạy cùng một mã nhiều lần, tôi sẽ nhận được các số ngẫu nhiên giống nhau

for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    returns = np.dot(weights, ind_er) # Lợi nhuận là tích số của lợi nhuận kỳ vọng riêng lẻ của tài sản và trọng số của nó
    p_ret.append(returns)
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum() # Phương sai danh mục đầu tư
    sd = np.sqrt(var) # Độ lệch chuẩn hàng ngày
    ann_sd = sd*np.sqrt(250) # Độ lệch chuẩn hàng năm = biến động
    p_vol.append(ann_sd)
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}
for counter, symbol in enumerate(stocks.columns.tolist()):
    data[symbol+' weight'] = [w[counter] for w in p_weights]
```

```python
portfolios  = pd.DataFrame(data)
portfolios.head() # Khung dữ liệu của 10000 danh mục đầu tư được tạo
```

```python
# Vẽ đường biên hiệu quả
portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[10,10])
```

DANH MỤC BIẾN ĐỘNG TỐI THIỂU (MINIMUM VOLATILITY PORTFOLIO)

```python
min_vol_port = portfolios.iloc[portfolios['Volatility'].idxmin()]
# idxmin() cung cấp cho tôi giá trị tối thiểu trong cột được chỉ định                              
min_vol_port
```

```python
# Vẽ sơ đồ danh mục biến động tối thiểu
plt.subplots(figsize=[10,10])
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
```

DANH MỤC RỦI RO TỐI ƯU (OPTIMAL RISKY PORTFOLIO)

```python
# Tìm danh mục rủi ro tối ưu
rf = 0.01 # Yếu tố rủi ro
optimal_risky_port = portfolios.iloc[((portfolios['Returns']-rf)/portfolios['Volatility']).idxmax()]
optimal_risky_port
```

```python
# Trích xuất trọng số từ danh mục rủi ro tối ưu
vcg_weight = optimal_risky_port['VCG weight']
ssi_weight = optimal_risky_port['SSI weight']
fpt_weight = optimal_risky_port['FPT weight']
dp3_weight = optimal_risky_port['DP3 weight']
frt_weight = optimal_risky_port['FRT weight']
vcb_weight = optimal_risky_port['VCB weight']

# Tên các tài sản
assets = ['VCB', 'SSI', 'FPT', 'DP3', 'FRT', 'VCG']

# Trọng số tương ứng
weights = [vcb_weight, ssi_weight, fpt_weight, dp3_weight, frt_weight, vcg_weight]

# Vẽ biểu đồ hình tròn như sau
fig, ax = plt.subplots(figsize=(8, 5))
pie = ax.pie(weights, labels=assets, autopct='%1.1f%%', startangle=140)
ax.legend(pie[0], assets, loc="center left", bbox_to_anchor=(1, 0.5))
plt.title('Phân bổ danh mục đầu tư')
plt.show()
```

```python
# Lập danh mục rủi ro tối ưu
plt.subplots(figsize=(10, 8))
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
plt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='g', marker='*', s=500)
```
